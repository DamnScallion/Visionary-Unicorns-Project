```python
import matplotlib.pyplot as plt
from ultralytics import YOLO
import os
import torch
import torch.nn as nn
```

```python
class CustomBinaryCrossEntropyLoss(nn.Module):
    def __init__(self):
        super(CustomBinaryCrossEntropyLoss, self).__init__()
        self.binary_cross_entropy = nn.BCEWithLogitsLoss()

    def forward(self, pred, target):
        # Assuming `pred` is the predicted output and `target` is the target tensor with binary labels
        loss = self.binary_cross_entropy(pred, target)
        return loss
```

```python
class CustomYOLOv8n(YOLO):
    def __init__(self, config_path):
        super(CustomYOLOv8n, self).__init__(config_path)
        self.loss_fn = CustomBinaryCrossEntropyLoss()

    def forward(self, x):
        return self.model(x)
```

```python
ROOT_DIR = '/content/Visionary-Unicorns-Project/yolo/config'
```

```python
model = CustomYOLOv8n('yolov8n.yaml')
```

```python
# Train the model with original given dataset for 10 epochs
results = model.train(data=os.path.join(ROOT_DIR, 'yolo_config.yaml'), epochs=100, imgsz=224, name='yolov8n_e10')
```
